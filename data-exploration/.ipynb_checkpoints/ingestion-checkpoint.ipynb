```python
import pandas as pd
# from sqlalchemy import create_engine
import psycopg2
import time
```

```python
# ingest click_streams
df_stream = pd.read_csv('../data/one-day-stream-filtered.csv')
df_stream.head()
```

```python
df_stream.columns
```

```python
df_stream[['product_id', 'quantity', 'item_price', 'promo_amount']] = \
    df_stream[['product_id', 'quantity', 'item_price', 'promo_amount']].fillna('0')
df_stream[['session_id', 'event_name', 'event_id', 'traffic_source',
           'payment_status', 'search_keywords', 'promo_code']] = \
    df_stream[['session_id', 'event_name', 'event_id', 'traffic_source',
               'payment_status', 'search_keywords', 'promo_code']].fillna('')
```

```python
df_stream = df_stream.astype({
    'session_id': str,
    'event_name': str,
    'event_id': str,
    'traffic_source': str,
    'product_id': float,
    'quantity': float,
    'item_price': float,
    'payment_status': str,
    'search_keywords': str,
    'promo_code': str,
    'promo_amount': float,
    'event_offset': float
})
```

```python
# transactions preparation
df_transac = pd.read_csv('../data/one-day-transactions-filtered.csv')
df_transac.head()
```

```python
df_transac.columns
```

```python
df_transac[['customer_id', 'promo_amount', 'shipment_fee', 'total_amount',
            'product_id', 'quantity', 'item_price']] = \
    df_transac[['customer_id', 'promo_amount', 'shipment_fee', 'total_amount',
                'product_id', 'quantity', 'item_price']].fillna('0')
df_transac[['booking_id', 'session_id', 'payment_method', 'payment_status',
            'promo_code', 'shipment_location_lat', 'shipment_location_long']] = \
    df_transac[['booking_id', 'session_id', 'payment_method', 'payment_status',
                'promo_code', 'shipment_location_lat', 'shipment_location_long']].fillna('')
```

```python
df_transac = df_transac.astype({
    'customer_id': float,
    'booking_id': str,
    'session_id': str,
    'payment_method': str,
    'payment_status': str,
    'promo_amount': float,
    'promo_code': str,
    'shipment_fee': float,
    'shipment_location_lat': str,
    'shipment_location_long': str,
    'total_amount': float,
    'product_id': float,
    'quantity': float,
    'item_price': float,
    'event_offset': float
})
```

```python
def insertion(df, query, pg_connection, connection_cursor):
    records = 0
    for i, row in df.iterrows():
        row_values = row.values.tolist()
        time.sleep(row_values[-1])
        print(f'Inserting a record after {row_values[-1]} seconds.')
        connection_cursor.execute(query, row_values[:-1])
        pg_connection.commit()
        records += connection_cursor.rowcount
        if i >= 2:
            break
    return records
    
```

```python
try:
    pg_connection = psycopg2.connect(
        user = 'user_a',
        password = 'pwd_a',
        host = '127.0.0.1',
        port = 5454,
        database = 'click_stream'
    )
    connection_cursor = pg_connection.cursor()
    # insert 5 clicks
    clicks_query = '''
    insert into clicks(session_id, event_name, event_id, traffic_source, product_id, quantity,
    item_price, payment_status, search_keywords, promo_code, promo_amount)
    values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    '''
    clicks_records = insertion(df_stream, clicks_query, pg_connection, connection_cursor)
    print(f'Newly inserted records in clicks table: {clicks_records}')

    # insert 5 transactions
    transactions_query = '''
    insert into transactions(customer_id, booking_id, session_id, payment_method,
       payment_status, promo_amount, promo_code, shipment_fee,
       shipment_location_lat, shipment_location_lon, total_amount,
       product_id, quantity, item_price)
    values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    '''
    transactions_records = insertion(df_transac, transactions_query, pg_connection, connection_cursor)
    print(f'Newly inserted records in transactions table: {transactions_records}')
    

except (Exception, psycopg2.Error) as e:
    print('Insertion error:', e)
finally:
    if pg_connection:
        connection_cursor.close()
        pg_connection.close()
        print('PG Connection closed.')
    
```
